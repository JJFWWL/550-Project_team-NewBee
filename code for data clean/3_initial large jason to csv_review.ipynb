```python
import json

def init_ds(json):
    ds= {}
    keys = json.keys()
    for k in keys:
        ds[k]= []
    return ds, keys

def read_json(file):
    dataset = {}
    keys = []
    with open(file) as file_lines:
        for count, line in enumerate(file_lines):
            data = json.loads(line.strip())
            if count ==0:
                dataset, keys = init_ds(data)
            for k in keys:
                dataset[k].append(data[k])
                
        return pd.DataFrame(dataset)
```

```python
yelp_reviews = read_json('../input/dataset-review/yelp_academic_dataset_review.json')
yelp_reviews.head()
```

```python
new_yelp_reviews=yelp_reviews.filter(['review_id','user_id','business_id','stars' ],axis=1)
new_yelp_reviews.head()
new_yelp_reviews.to_csv('./new_yelp_reviews.csv',index=False)
```
